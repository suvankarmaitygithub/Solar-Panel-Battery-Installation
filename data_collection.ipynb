```python
import pandas as pd
import numpy as np

# Simulate data
np.random.seed(42)
data = pd.DataFrame({
    'household_id': range(1, 101),
    'grid_reliability': np.random.uniform(0.8, 1.0, 100),  # 0 to 1 scale
    'avg_power_outage_duration': np.random.uniform(0, 5, 100),  # in hours per month
    'peak_power_requirement': np.random.uniform(2, 10, 100),  # in kW
    'off_peak_power_requirement': np.random.uniform(1, 5, 100),  # in kW
    'solar_generation_capacity': np.random.uniform(3, 12, 100),  # in kW
    'excess_electricity': np.random.uniform(0, 6, 100),  # in kWh
    'battery_capacity': np.random.choice([5, 10, 15], 100),  # in kWh
    'battery_type': np.random.choice(['Lithium-ion', 'Lead-acid'], 100),
    'purchase_battery': np.random.choice([0, 1], 100)  # 0: No, 1: Yes
})

data.to_csv('prosumer_data.csv', index=False)

```
